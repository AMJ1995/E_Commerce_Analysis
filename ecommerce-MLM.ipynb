```python
def reduce_mem_usage(df, verbose=True):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    start_mem = df.memory_usage().sum() / 1024**2
    for col in df.columns:
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)

    end_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage after optimization is: {:.2f} MB'.format(end_mem))
    print('Decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))

    return df
```

# Conversion Rates

As I mentioned before, we have three type of events view, add to cart and purchase. As you may have noticed from the graph above, not every user who views a product adds that item to their cart and is even less likely to purchase it. Most users tend to have look at the product and its price before moving forward in their decision making process.

This is where conversion rates can be used to give us an idea of how many users actually purchased the product as opposed to how many times the products are viewed or added to the cart.

How many products are actually purchased as opposed to number of products added to the cart?

We are trying to determine the following:
* Count of people who viewed the item.

# Preparing data for a Logisitic Regression Model

In order to predict whether the product added to the cart is actually purchased by the customer based on factors such as its category, event_weekday, and activity of the user in that session etc

We need new features in training data

*   category - category
*   sub_category - subcategory
*   event_weekday - weekday of the event
*   activity_count - number of activity in that session
*   brand - brand the product belongs to
*   is_purchased - whether the put in cart item is purchased.

The training dataset contains every non-duplicated cart transactions with above mentioned features. We will use these features with their original price and brand to predict whether the customer will eventually purchase the item in the cart.

#List of people who has bought or added products to the cart

```python
#List of people who has bought or added products to the cart
cart_purchase_users = df.loc[df["event_type"].isin(["cart","purchase"])].drop_duplicates(subset=['user_id'])
cart_purchase_users.dropna(how='any', inplace=True)
print(cart_purchase_users)
```

### All activities of above users view events

```python
cart_purchase_users_all_activity = df.loc[df['user_id'].isin(cart_purchase_users['user_id'])]
print(cart_purchase_users_all_activity)
```

### Counting the number of activities in one user session

```python

activity_in_session = cart_purchase_users_all_activity.groupby(['user_session'])['event_type'].count().reset_index()
activity_in_session = activity_in_session.rename(columns={"event_type":"activity_count"})
print(activity_in_session)
```

### Extract the event_date from event_time column and find on which date the activity occurs

```python
def convert_time_to_date(utc_timestamp):
  utc_date = datetime.strptime(utc_timestamp[0:10],'%Y-%m-%d').date()
  return utc_date
```

```python
df['event_date'] = df['event_time'].apply(lambda s:convert_time_to_date(s))
```

### Splitting the category and sub-category through string handling

```python
df_targets = df.loc[df["event_type"].isin(["cart","purchase"])].drop_duplicates(subset = ['event_type','product_id','price','user_id','user_session'])
df_targets["is_purchased"] = np.where(df_targets["event_type"]=="purchase",1,0)
df_targets["is_purchased"] = df_targets.groupby(["user_session","product_id"])["is_purchased"].transform("max")
df_targets = df_targets.loc[df_targets["event_type"]=='cart'].drop_duplicates(["user_session","product_id","is_purchased"])
df_targets['event_weekday'] = df_targets['event_date'].apply(lambda s:s.weekday())
df_targets.dropna(how = 'any',inplace = True)
df_targets["category"] = df_targets["category_code"].str.split(".",expand = True)[0].astype('category')
df_targets["sub_category"] = df_targets["category_code"].str.split(",",expand = True)[0].astype('category')
```

```python
df_targets = df_targets.merge(activity_in_session,on = 'user_session',how ='left')
df_targets['activity_count'] = df_targets['activity_count'].fillna(0)
df_targets.head()
```

```python
df_targets = df_targets.drop(columns=['event_time'])
df_targets = df_targets.drop(columns=['product_id'])
df_targets = df_targets.drop(columns=['category_id'])
df_targets = df_targets.drop(columns=['category_code'])
df_targets = df_targets.drop(columns=['user_id'])
df_targets = df_targets.drop(columns=['user_session'])
```

### Resampling data to have equal number of purchased and not purchased items

Number of rows when the item was purchased was around 5 and not purchased was around 8.
