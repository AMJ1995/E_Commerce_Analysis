```python
import pandas as pd
import matplotlib as plt
import numpy as np
import glob
import os
```

```python
data0 = pd.read_csv("big_split/chunk0.csv")
```

```python
data1 = pd.read_csv("big_split/chunk1.csv")
data2 = pd.read_csv("big_split/chunk2.csv")
```

```python
data3 = pd.read_csv("big_split/chunk3.csv")
data4 = pd.read_csv("big_split/chunk4.csv")
```

```python
data5 = pd.read_csv("big_split/chunk5.csv")
data6 = pd.read_csv("big_split/chunk6.csv")
```

```python
data0 = data0.drop(columns=['user_id'])
data0 = data0.drop(columns=['user_session'])
```

```python
data1 = data1.drop(columns=['user_id'])
data1 = data1.drop(columns=['user_session'])
data2 = data2.drop(columns=['user_id'])
data2 = data2.drop(columns=['user_session'])
```

```python
data3 = data3.drop(columns=['user_id'])
data3 = data3.drop(columns=['user_session'])
data4 = data4.drop(columns=['user_id'])
data4 = data4.drop(columns=['user_session'])
data5 = data5.drop(columns=['user_id'])
data5 = data5.drop(columns=['user_session'])
data6 = data6.drop(columns=['user_id'])
data6 = data6.drop(columns=['user_session'])
```

```python
data0.dropna(how = 'any', inplace = True)
```

```python
data1.dropna(how = 'any', inplace = True)
data2.dropna(how = 'any', inplace = True)
data3.dropna(how = 'any', inplace = True)
data4.dropna(how = 'any', inplace = True)
data5.dropna(how = 'any', inplace = True)
data6.dropna(how = 'any', inplace = True)
```

```python
purchase_data0 = data0[data0['event_type'] == 'purchase']
purchase_data0 = purchase_data0.drop(columns=['event_type'])
purchase_data1 = data1[data1['event_type'] == 'purchase']
purchase_data1 = purchase_data1.drop(columns=['event_type'])
purchase_data2 = data2[data2['event_type'] == 'purchase']
purchase_data2 = purchase_data2.drop(columns=['event_type'])
purchase_data2.head()
```

```python
purchase_data3 = data3[data3['event_type'] == 'purchase']
purchase_data3 = purchase_data3.drop(columns=['event_type'])
purchase_data4 = data4[data4['event_type'] == 'purchase']
purchase_data4 = purchase_data4.drop(columns=['event_type'])
purchase_data5 = data5[data5['event_type'] == 'purchase']
purchase_data5 = purchase_data5.drop(columns=['event_type'])
purchase_data6 = data6[data6['event_type'] == 'purchase']
purchase_data6 = purchase_data6.drop(columns=['event_type'])
purchase_data6.head()
```

```python
cart_data0 = data0[data0['event_type'] == 'cart']
cart_data0 = cart_data0.drop(columns=['event_type'])
cart_data1 = data1[data1['event_type'] == 'cart']
cart_data1 = cart_data1.drop(columns=['event_type'])
cart_data2 = data2[data2['event_type'] == 'cart']
cart_data2 = cart_data2.drop(columns=['event_type'])
cart_data3 = data3[data3['event_type'] == 'cart']
cart_data3 = cart_data3.drop(columns=['event_type'])
cart_data4 = data4[data4['event_type'] == 'cart']
cart_data4 = cart_data4.drop(columns=['event_type'])
cart_data5 = data5[data5['event_type'] == 'cart']
cart_data5 = cart_data5.drop(columns=['event_type'])
cart_data6 = data6[data6['event_type'] == 'cart']
cart_data6 = cart_data6.drop(columns=['event_type'])
cart_data6.head()
```

```python
view_data0 = data0[data0['event_type'] == 'view']
view_data0 = view_data0.drop(columns=['event_type'])
view_data1 = data1[data1['event_type'] == 'view']
view_data1 = view_data1.drop(columns=['event_type'])
view_data2 = data2[data2['event_type'] == 'view']
view_data2 = view_data2.drop(columns=['event_type'])
view_data3 = data3[data3['event_type'] == 'view']
view_data3 = view_data3.drop(columns=['event_type'])
view_data4 = data4[data4['event_type'] == 'view']
view_data4 = view_data4.drop(columns=['event_type'])
view_data5 = data5[data5['event_type'] == 'view']
view_data5 = view_data5.drop(columns=['event_type'])
view_data6 = data6[data6['event_type'] == 'view']
view_data6 = view_data6.drop(columns=['event_type'])
view_data6.head()
```

```python
view_data = [view_data0, view_data1, view_data2, view_data3, view_data4, view_data5, view_data6]
view_df = pd.concat(view_data)
```

```python
purchase_data = [purchase_data0, purchase_data1, purchase_data2, purchase_data3, purchase_data4, purchase_data5, purchase_data6]
purchase_df = pd.concat(purchase_data)
```

```python
cart_data = [cart_data0, cart_data1, cart_data2, cart_data3, cart_data4, cart_data5, cart_data6]
cart_df = pd.concat(cart_data)
```
