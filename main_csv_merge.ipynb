```python
import pandas as pd
import matplotlib as plt
import numpy as np
```

```python
nov_data = pd.read_csv('../e_commerce_analysis_temp/archive/2019-Nov.csv')
nov_data.info()
```

```python
oct_data = pd.read_csv('../e_commerce_analysis_temp/archive/2019-Oct.csv')
```

```python
nov_data.dropna(how = 'any', inplace = True)
oct_data.dropna(how = 'any', inplace = True)
```

```python
main_data = pd.concat([oct_data,nov_data])
main_data.head()
```

```python
main_data.tail()
```

```python
# Adding index for SQL Primary Key
main_data['event_index'] = np.arange(1, main_data.shape[0] + 1)
main_data = main_data[['event_index','event_time','event_type','product_id','category_id','category_code','brand','price','user_id','user_session']]
main_data.head()
```

```python
main_data.tail()
```

```python
main_data.to_csv('../e_commerce_analysis_temp/csv/main_2019.csv',encoding='utf-8',index=False)
```
